```python
from collections import defaultdict

### HGT path:
hgt_node = 172
hgt_path = [hgt_node]
while hgt_node <= 204:
    hgt_node = tree.parent(hgt_node)
    hgt_path.append(hgt_node)

#print(hgt_path)

children = defaultdict(list)
for leaf in range(num_samples):
    node = leaf
    while node < tree.root and len(children[node]) < 2:
        parent = tree.parent(node)
        children[parent].append(node)
        node = parent
        if parent == tskit.NULL:
            break


for node in sorted(list(children)):
    if len(children[node]) == 1:
        parent = tree.parent(node)
        if parent == tskit.NULL:
            continue  # kein Parent → überspringen
        child = children[node][0]  # das eine Kind

        # 1. altes Kind durch neues Kind ersetzen
        if node in children[parent]:
            children[parent].remove(node)
        children[parent].append(child)

        # 2. entferne den Zwischenknoten
        del children[node]


for node in reversed(hgt_path):
    print(node, tree.children(node), children[node], list(tree.leaves(node)))
```

```python
import glob
import os
import random
import h5py
import pickle
import networkx as nx
import matplotlib.pyplot as plt

output_dir = r"C:\Users\uhewm\Desktop\ProjectHGT\simulation_chunks"
all_files = sorted(glob.glob(os.path.join(output_dir, "*.h5")))

for file in random.sample(all_files, 1):
    with h5py.File(file, "r") as f:
        grp = f["results"]
        graph_properties = pickle.loads(grp["graph_properties"][()])
        # Tensors erstellen
        nodes = torch.tensor(graph_properties[0])          # Shape [199]
        edges = torch.tensor(graph_properties[1])          # Shape [2, 198]
        coords = torch.tensor(graph_properties[2].T)       # Shape [2, 199]

# Erstelle einen gerichteten Graphen
G = nx.DiGraph()

# Füge Knoten hinzu (optional mit Koordinaten als Attribut)
for i, node_id in enumerate(nodes.tolist()):
    G.add_node(node_id, core_distance = coords[:, i].tolist()[0], allele_distance = coords[:, i].tolist()[1])

# Füge Kanten hinzu
edge_list = edges.tolist()
for src, dst in zip(edge_list[0], edge_list[1]):
    G.add_edge(src, dst)

```

```python
def hierarchy_pos(G, root=None, width=1.0, vert_gap=0.2, vert_loc=0, xcenter=0.5):
    """
    Positioniert die Knoten eines gerichteten Graphen G in einer hierarchischen Struktur.
    Quelle: modifizierte Version basierend auf Joel's answer auf StackOverflow
    """
    def _hierarchy_pos(G, root, width=1.0, vert_gap=0.2, vert_loc=0, xcenter=0.5, pos=None, parent=None):
        if pos is None:
            pos = {root: (xcenter, vert_loc)}
        else:
            pos[root] = (xcenter, vert_loc)
        neighbors = list(G.successors(root))
        if parent is not None and parent in neighbors:
            neighbors.remove(parent)
        if len(neighbors) != 0:
            dx = width / len(neighbors)
            nextx = xcenter - width/2 - dx/2
            for neighbor in neighbors:
                nextx += dx
                pos = _hierarchy_pos(G, neighbor, width=dx, vert_gap=vert_gap,
                                     vert_loc=vert_loc - vert_gap, xcenter=nextx, pos=pos, parent=root)
        return pos

    if root is None:
        # Versuche Wurzel automatisch zu finden
        root = [n for n, d in G.in_degree() if d == 0][0]

    return _hierarchy_pos(G, root, width, vert_gap, vert_loc, xcenter)

pos = hierarchy_pos(G, root = 198)
plt.figure(figsize=(12, 8))
nx.draw(G, pos,
        node_color="lightblue",
        node_size=2000,
        arrows=True,
        with_labels=False)

node_labels = {
    n: f"{n}\n({round(G.nodes[n]['core_distance'], 2)}, {round(G.nodes[n]['allele_distance'], 2)})"
    for n in G.nodes
}
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10)

plt.title("Hierarchisches Baum-Layout mit Features")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
def hierarchy_pos(G, root=None, width=1., vert_gap=0.2, vert_loc=0, x_center=0.5):
    pos = {}

    def _hierarchy_pos(G, node, x, y, dx):
        children = list(G.successors(node))
        if not children:
            pos[node] = (x, y)
        else:
            nextx = x - dx * (len(children) - 1) / 2
            for child in children:
                _hierarchy_pos(G, child, nextx, y - vert_gap, dx / max(len(children), 1))
                nextx += dx
            pos[node] = (x, y)

    _hierarchy_pos(G, root, x_center, vert_loc, width)
    return pos

def hierarchy_pos_all_roots(G, roots=None, vert_gap=0.2, vert_loc=0, x_start=0.0):
    pos = {}
    if roots is None:
        roots = [n for n, d in G.in_degree() if d == 0]

    x_offset = x_start
    for root in roots:
        subtree = nx.descendants(G, root) | {root}
        subG = G.subgraph(subtree)
        pos_sub = hierarchy_pos(subG, root=root, vert_gap=vert_gap, vert_loc=vert_loc, x_center=x_offset)

        # Max X-Koordinate dieses Subbaums bestimmen
        max_x = max(p[0] for p in pos_sub.values()) if pos_sub else 0
        width = max_x - x_offset if max_x > x_offset else 1.0  # fallback

        # aktualisieren
        pos.update(pos_sub)
        x_offset = max_x + 1.0  # Abstand zwischen Bäumen

    return pos


# Hole Positionen für alle Teilbäume im Graphen (falls mehrere Wurzeln)
pos = hierarchy_pos_all_roots(G, vert_gap=0.3)

plt.figure(figsize=(12, 8))
nx.draw(G, pos,
        node_color="lightblue",
        node_size=2000,
        arrows=True,
        with_labels=False)

# Features als Labels
node_labels = {
    n: f"{n}\n({round(G.nodes[n].get('core_distance', 0), 2)}, {round(G.nodes[n].get('allele_distance', 0), 2)})"
    for n in G.nodes
}
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10)

plt.title("Hierarchisches Baum-Layout mit Features")
plt.axis("off")
plt.tight_layout()
plt.show()
```

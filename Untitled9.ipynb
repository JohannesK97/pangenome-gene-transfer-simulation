```python
from collections import defaultdict
import tskit

def simplify_tree(tree, root, num_samples):
    
    children = defaultdict(list)
    for leaf in range(num_samples):
        node = leaf
        while node < root and len(children[node]) < 2:
            parent = tree.parent(node)
            children[parent].append(node)
            node = parent
            if parent == tskit.NULL:
                break
                  
    for node in list(children):
        if len(children[node]) == 1:
            parent = tree.parent(node)
            if parent == tskit.NULL:
                continue  # kein Parent → überspringen
            child = children[node][0]  # das eine Kind
    
            # 1. altes Kind durch neues Kind ersetzen
            if node in children[parent]:
                children[parent].remove(node)
            children[parent].append(child)
    
            # 2. entferne den Zwischenknoten
            del children[node]

    parents = defaultdict(int)
    for parent, child_list in children.items():
        for child in child_list:
            parents[child] = parent

    node_to_leaf = defaultdict(list)
    for child in sorted(list(parents)):
        if child < num_samples:
            node_to_leaf[child] = [child]
        node_to_leaf[parents[child]].extend(node_to_leaf[child])
    
    return children, parents, node_to_leaf

def build_directed_graph(tree, gene_absence_presence_vector: np.ndarray, alleles_list: np.ndarray, clonal_root_node: int):

    if gene_absence_presence_vector.ndim != 1:
        raise ValueError("gene_presence_absence_vector muss ein 1D-Vektor sein (nur eine Zeile).")

    num_samples = len(gene_absence_presence_vector)

    simplified_mts_children, simplified_mts_parents, node_to_leaf = simplify_tree(tree, clonal_root_node, num_samples) 

    G = nx.DiGraph()
    for parent, child_list in simplified_mts_children.items():
        for child in child_list:
            G.add_edge(child, parent)
    
    for node in G.nodes:
        G.nodes[node]["core_distance"] = 0
        G.nodes[node]["allele_distance"] = 0

    ### Add clonal distances:
    for node in range(clonal_root_node): # Going up starting from the leaves.
        if node < num_samples and gene_absence_presence_vector[node] == 1:
            parent = simplified_mts_parents[node]
            G.nodes[parent]["core_distance"] += tree.get_time(parent)
        elif node >= num_samples and node in G.nodes and G.nodes[node]["core_distance"] > 0:
            parent = simplified_mts_parents[node]
            G.nodes[parent]["core_distance"] += tree.get_time(parent) - tree.get_time(node) + G.nodes[node]["core_distance"]

    ### Add allele distances:
    allele_distances = defaultdict(list)
    
    for node, leaves in node_to_leaf.items():
        subset = alleles_list[:, leaves]
        distance = np.count_nonzero(np.ptp(subset, axis=1))  # ptp = max - min
        allele_distances[tuple(leaves)] = distance
    
    for node in list(simplified_mts_parents) + [clonal_root_node]:
        G.nodes[node]["allele_distance"] = allele_distances[tuple(node_to_leaf[node])]

    return G

tree = mts.first()
simplified_mts_children, simplified_mts_parents, node_to_leaf = simplify_tree(tree, clonal_root_node, num_samples)  
node_to_leaf
```

```python
import networkx as nx
import tskit
import torch
from torch_geometric.utils import from_networkx

tree = core_tree.first()

G = nx.DiGraph()

for node in tree.nodes():
    G.add_node(node)

for node in tree.nodes():
    parent = tree.parent(node)
    if parent != tskit.NULL:
        G.add_edge(node, parent)  # gerichtete Kante von Child zu Parent

features = {}

for node in G.nodes:
    feature_1 = 0
    feature_2 = 1
    features[node] = np.array([feature_1, feature_2])

# Features in Graph speichern (z. B. als Attribut)
nx.set_node_attributes(G, values=features, name="features")

# Umwandeln in PyG Data-Objekt
pyg_graph = from_networkx(G)

# Stelle sicher, dass du alle originalen Node-IDs in richtiger Reihenfolge hast
node_ids_sorted = sorted(G.nodes())

# Baue Feature-Matrix in genau dieser Reihenfolge
feature_tensor = torch.tensor(
    [features[node_id] for node_id in node_ids_sorted],
    dtype=torch.float
)

# Setze Feature-Matrix und speichere die Zuordnung
pyg_graph.x = feature_tensor
pyg_graph.node_id_map = node_ids_sorted  # <- das kannst du später verwenden

#pyg_graph.x = feature_tensor
```
